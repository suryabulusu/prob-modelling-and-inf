```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.io
```

```python
# Loading mat files - https://stackoverflow.com/questions/874461/read-mat-files-in-python
file = scipy.io.loadmat("mnist_small.mat")
X = file["X"]
y = file["Y"]
N, D = np.shape(X)

#normalize -> 0 to 1 range
X = X / 255
# randomize X and y
```

```python
idx = np.random.randint(10000, size = 10000)
X = X[idx]
y = y[idx]
T = 2
# K = 5, 10, 15, 20. Please change the value of K here
K = 15
n = N
X_ = X[:n]
y_ = y[:n]
mu = np.random.rand(K, D)
p = np.ones(K)/K
resp = np.empty([n, K])
var = 1
cll = []
for t in range(T):
    print("T is", t)
    # E-Step
    for i in range(n):
        exps = np.linalg.norm(mu - X_[i], axis = 1)**2 / (2*var)
        lognum = np.log(p) - exps
        minexp = np.min(exps)
        logden = np.log(p @ np.exp(minexp - exps)) - minexp
        resp[i] = np.exp(lognum-logden)
    
    # M-Step
    var = 0
    for k in range(K):
        N_k = np.sum(resp[:, k])
        p[k] = N_k / n 
        mu[k] = X_.T @ resp[:, k] / (N_k + 1e-5)
        var += np.linalg.norm(X_ - mu[k], axis=1)**2 @ resp[:, k]
    var = (1/(n*D))*var
    
    # compute cll
    temp = 0
    for k in range(K):
        temp += (resp[:, k] @ (np.log(p[k]) -0.5*np.linalg.norm(X_ - mu[k], axis = 1)**2/var))
        temp -= (D/2)*np.log(var)*np.sum(resp[:,k])
    cll.append(temp)
    
plt.plot(range(T), cll)
plt.title("CLL for K = %i"% K)
plt.show()

for k in range(K):
    plt.figure(figsize=(2,2))
    plt.imshow(np.reshape(mu[k], [28,28]))
    plt.show()
```
