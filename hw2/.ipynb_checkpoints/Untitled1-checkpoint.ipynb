```python
B=100
T = 100
np.random.seed(7)
#K = [5, 10, 15, 20]
K = 20
mu = np.random.rand(K, D)
pi = np.ones(K)/K
resp = np.empty([B, K])
var = 1
kappa = 0.55
for t in range(T):
    if (t % 10 == 0): print("T is", t)
    neta=(1+t)**(-kappa)
    y_ = np.asarray([])
    for i in range(10):
        yidx = np.where(y == i)[0]
        yidx = yidx[np.random.randint(yidx.shape[0], size = 10)]
        y_ = np.append(y_, yidx).astype(int)
        
    idx_data = np.random.randint(100, size = 100)
        
    X_ = X[y_[idx_data]]
    
    for i in range(B):
        exps = np.linalg.norm(mu - X_[i], axis = 1)**2 / (2*var)
        lognum = np.log(pi) - exps
        maxexp = np.max(lognum)
        sum1 = np.sum([np.exp(lognum-maxexp)])
        resp[i] = np.exp(lognum - maxexp  - np.log(sum1)) 
    
    for k in range(K):
        N_k = np.sum(resp[:, k])
        pi[k] = (1-neta)*pi[k] + neta*N_k / B
        sum3=np.zeros((1,D))
        for n in range(B):
            sum3=sum3+resp[n][k]*X_[n]
        if(sum2!=0):
            mu[k]=neta*sum3/sum2+(1-neta)*mu[k]
    
    #pi = (1-neta)*pi + neta*pi
    
    sum4=0
    for n in range(B):
        for k in range(K):
            sum4=sum4+resp[n][k]*np.square(np.linalg.norm(X_[n]-mu[k]))
    var=neta*sum4/(B*D)+(1-neta)*var
```
