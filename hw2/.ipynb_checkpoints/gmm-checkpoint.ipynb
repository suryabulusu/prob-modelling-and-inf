```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.io
```

```python
# Loading mat files - https://stackoverflow.com/questions/874461/read-mat-files-in-python
file = scipy.io.loadmat("mnist_small.mat")
X = file["X"]
y = file["Y"]
N, D = np.shape(X)

#normalize -> 0 to 1 range
X = X / 255
# randomize X and y
```

```python
idx = np.random.randint(10000, size = 10000)
X = X[idx]
y = y[idx]
T = 200
#K = [5, 10, 15, 20]
K = 15
n = N
X_ = X[:n]
y_ = y[:n]
mu = np.random.rand(K, D)
p = np.ones(K)/K
resp = np.empty([n, K])
var = 1
cll = []
for t in range(T):
    if (t % 10 == 0): print("T is", t)
    # E-Step
    for i in range(n):
        exps = np.linalg.norm(mu - X_[i], axis = 1)**2 / (2*var)
        lognum = np.log(p + 1e-5) - exps
        minexp = np.min(exps)
        logden = np.log(p @ np.exp(minexp - exps)) - minexp
        resp[i] = np.exp(lognum-logden)
    
    # M-Step
    var = 0
    for k in range(K):
        N_k = np.sum(resp[:, k])
        p[k] = N_k / n 
        mu[k] = X_.T @ resp[:, k] / (N_k + 1e-5)
        var += np.linalg.norm(X_ - mu[k], axis=1)**2 @ resp[:, k]
    var = (1/(n*D))*var
    
    # compute cll
    temp = 0
    for k in range(K):
        temp += (resp[:, k] @ (np.log(p[k] + 1e-5) -0.5*np.linalg.norm(X_ - mu[k], axis = 1)**2/var))
        temp -= (D/2)*np.log(var)*np.sum(resp[:,k])
    cll.append(temp)
```

```python
plt.plot(range(T), cll)
plt.show()
```

```python
for k in range(K):
    plt.figure(figsize=(2,2))
    plt.imshow(np.reshape(mu[k], [28,28]))
    plt.show()
```

```python
# Online EM
T = 150
#K = [5, 10, 15, 20]
K = 15
n = 100
mu = np.random.rand(K, D)
p = np.ones(K)/K
resp = np.empty([n, K])
var = 1
mu_old, p_old, var_old = mu, p, var
cll = []
kap = 0.55
flag = 0
t_l = 0 
for t in range(T):
    if (t % 10 == 0): print("T is", t)
    lr = (1+t_l)**(-1*kap)    
    
    # generate data; ensures each batch has 10 examples from each digit
    y_ = np.asarray([])
    for i in range(10):
        yidx = np.where(y == i)[0]
        yidx = yidx[np.random.randint(yidx.shape[0], size = 10)]
        y_ = np.append(y_, yidx).astype(int)
        
    idx_data = np.random.randint(100, size = 100)
        
    X_ = X[y_[idx_data]]
    
    # E-Step
    for i in range(n):
        exps = np.linalg.norm(mu - X_[i], axis = 1)**2 / (2*var)
        lognum = np.log(p) - exps
        minexp = np.min(exps)
        logden = np.log(p @ np.exp(minexp - exps)) - minexp
        resp[i] = np.exp(lognum-logden)
        
    # M-Step
    var = 0
    for k in range(K):
        N_k = np.sum(resp[:, k])
        if (N_k < 1e-5):
            # Ignore batch if N_k is too small :(
            flag = 1
            continue
        p[k] = N_k / n 
        if (N_k != 0):
            mu[k] = X_.T @ resp[:, k] / N_k
        var += np.linalg.norm(X_ - mu[k], axis=1)**2 @ resp[:, k]
    
    if (flag == 1):
        flag = 0
        # ignore batch 
        continue
        
    var = (1/(n*D))*var
    
    mu = mu*lr + mu_old*(1-lr)
    p = p*lr + p_old*(1-lr)
    var = var*lr + var_old*(1-lr)
    mu_old, p_old, var_old = mu, p, var
    
    t_l += 1 # actual iterations (not accounting for ignored batches)
    
    # compute cll
    temp = 0
    for k in range(K):
        temp += (resp[:, k] @ (np.log(p[k] + 1e-5) -0.5*np.linalg.norm(X_ - mu[k], axis = 1)**2/var))
        temp -= (D/2)*np.log(var)*np.sum(resp[:,k])
    cll.append(temp)
```
